### Cleaning Data

```python
#Combine year, month, day, and hour into a single datetime column for each dataset
for df in [aotizhongxin_df, changping_df, tiantan_df]:
    df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
```

```python
#Fill missing values
#For numerical columns, fill missing values with the median of each column
#For 'wd' (wind direction), fill missing values with 'Unknown'
columns_to_fill_median = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']
for df in [aotizhongxin_df, changping_df, tiantan_df]:
    for column in columns_to_fill_median:
        df[column] = df[column].fillna(df[column].median())
    df['wd'] = df['wd'].fillna('Unknown')
```

```python
#Drop unnecessary columns (year, month, day, hour) after creating the datetime column
for df in [aotizhongxin_df, changping_df, tiantan_df]:
    df.drop(columns=['year', 'month', 'day', 'hour'], inplace=True)
```

**Insight:**
- By combining the year, month, day, and hour into a single datetime column, the datasets have been significantly enhanced for time-series analysis. This transformation facilitates easier querying and visualization of air quality trends over time, allowing for more effective monitoring and assessment of pollution levels.
- The filling of missing values with the median for numerical columns and categorizing wind direction as 'Unknown' improves the overall quality of the datasets. This step reduces the impact of missing data on analyses and ensures that subsequent analyses (such as statistical evaluations or visualizations) yield more accurate and reliable insights into air quality patterns.

## Exploratory Data Analysis (EDA)

### Explore ...

```python
#Combine the datasets by concatenating them
df_combined = pd.concat([aotizhongxin_df, changping_df, tiantan_df])
```

```python
#Plot the monthly trend for PM2.5
df_combined['datetime'] = pd.to_datetime(df_combined['datetime'])
df_combined.set_index('datetime', inplace=True)
monthly_avg_pm25 = df_combined['PM2.5'].resample('M').mean()

plt.figure(figsize=(10,6))
plt.plot(monthly_avg_pm25, color='b', label='PM2.5 (Monthly Avg)')
plt.title('Monthly Average PM2.5 Levels')
plt.xlabel('Date')
plt.ylabel('PM2.5 Concentration')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Sort the combined data by PM2.5 levels to explore the top 5 highest and lowest pollution levels
df_combined['datetime'] = pd.to_datetime(df_combined['datetime'])
df_combined_sorted = df_combined.sort_values(by="PM2.5", ascending=False)

top_5 = df_combined_sorted[['station', 'datetime', 'PM2.5']].head()
bottom_5 = df_combined_sorted[['station', 'datetime', 'PM2.5']].tail()

print("5 Highest PM2.5 levels:", top_5)
print("\n5 Lowest PM2.5 levels:", bottom_5)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Is there any relationship between wind speed and pollutant concentration?

```python
sns.scatterplot(x='WSPM', y='PM2.5', data=aotizhongxin_df)
plt.title('Wind Speed vs PM2.5 - Aotizhongxin')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('PM2.5 Concentration')
plt.show()
```

### Pertanyaan 2: What is the distribution of atmospheric pressure (PRES) and rainfall (RAIN) data?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PRES', y='PM2.5', data=aotizhongxin_df)
plt.title('Atmospheric Pressure vs PM2.5 - Aotizhongxin')
plt.xlabel('Atmospheric Pressure (hPa)')
plt.ylabel('PM2.5 Concentration')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='RAIN', y='PM2.5', data=aotizhongxin_df)
plt.title('Rainfall vs PM2.5 - Aotizhongxin')
plt.xlabel('Rainfall (mm)')
plt.ylabel('PM2.5 Concentration')
plt.show()
```

## Conclusion

- Conclution pertanyaan 1: Wind speed (WSPM) can influence the dispersion of pollutants. Higher wind speeds tend to spread air pollution, potentially lowering its concentration.
- Conclution pertanyaan 2:  Weather conditions like atmospheric pressure and rainfall can significantly influence air quality. For instance, rain can wash away pollutants from the air.
